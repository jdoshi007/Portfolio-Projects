```python
#import libraries
import pandas as pd
import seaborn as sns 
import numpy as np


import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) 


#Read the data


df = pd.read_csv(r"C:\Users\jashd\Downloads\archive (1)\movies.csv")


```

```python
df.head()
```

```python
#missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))



```

```python
#Data types
df.dtypes
```

```python

#changing data type

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
del df['yearcorrect']
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Budget
# scatter plot budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')




plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  # Adjust figure size

sns.regplot(x="gross", y="budget", data=df,  
            scatter_kws={"color": "red", "alpha": 0.5},  # Adjust transparency
            line_kws={"color": "blue", "linewidth": 2})  # Make line more visible

plt.title("Budget vs. Gross Earnings", fontsize=14)  
plt.xlabel("Gross Earnings", fontsize=12)  
plt.ylabel("Budget", fontsize=12)  
plt.grid(True)  # Add grid for better readability

plt.show()
```

```python

# Ensure df has only numeric values
df_numeric = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = df_numeric.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1, linecolor="black")
plt.title("Correlation Matrix for Numeric Feature", fontsize=14)
plt.xlabel("Movie Features", fontsize=12)
plt.ylabel("Movie Features", fontsize=12)
plt.show()

```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

#df_numerized
```

```python

# Ensure df has only numeric values
df_numeric = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = df_numerized.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1, linecolor="black")
plt.title("Correlation Matrix for Numeric Feature", fontsize=14)
plt.xlabel("Movie Features", fontsize=12)
plt.ylabel("Movie Features", fontsize=12)
plt.show()
```
